```python
#cell 1:
import numpy as np
import matplotlib.pyplot as plt

Age_groups = ["6-12", "13-18", "19-25"]
ballet_count = [10, 15, 5]
hiphop_counts = [5, 20, 10]
x = np.arange(len(Age_groups))
width = 0.35

plt.bar(x - width/2, ballet_count, width, label = "ballet", color='r')
plt.bar(x + width/2, hiphop_counts, width, label='hip_hop', color= 'y')

plt.xlabel("Age Group")
plt.ylabel("number of participants")
plt.title("dance club participation by age group")
plt.xticks(x, Age_groups)
plt.legend()
plt.show()
 

```

```python
#cell3:
x = np.linspace(0, 2*np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label="sin(x)")
plt.title("sin(x)")
plt.show()

plt.plot(x, y_sin, label="cos(x)", c="g")
plt.title("cos(x)")
plt.show()

plt.plot(x, y_sin, label="sin(x)")
plt.plot(x, y_cos, label="cos(x)", c="g")
plt.title("sin & cos")
plt.legend()
plt.show()
```
